```python
import requests
```

세션 생성

```python
#세션 객체 생성
with requests.Session()as session:
    #세션 객체에 웹 브라우저 정보(USERAGENT ) 주입
    session.headers.update({
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
        })
    
    url = "https://data.hossam.kr/py/sample.txt" #요청할 데이터의 url
    r=session.get(url) #요청 결과 받아오기-->http get 요청

    #HTTP상태값이 200이 아닌경우는 강제로 에러를 발생시켜서 코드의 진행을 중단시킴
    if r.status_code !=200:
        msg = "[%d Error] %s 에러가 발생함" % (r.status_code,r.reason)
        raise Exception(msg)
    
    print(r)
```

```python
# 수신된 데이터의 인코딩 설정 ( 한글이 깨진다면 euc-kr로 변경)
r.encoding = "utf-8"

#수신된 결과 확인 --> 웹에서 가져온 모든 데이터는 문자열 형식임
print(r.text)
```

```python
with requests.Session()as session:
    session.headers.update({
        "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
        })
    url = "https://data.hossam.kr/py/sample.json"
    r=session.get(url)

    if r.status_code !=200:
        msg = "[%d Error]%s 에러가 발생함" %(r.status_code, r.reason)
        raise Exception(msg)
    
    print(r)
```
