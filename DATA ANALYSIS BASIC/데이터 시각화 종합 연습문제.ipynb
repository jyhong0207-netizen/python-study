
수명에 대해 기술통계량을 구하고 상자그림을 그려라. 최종적으로 분석 결과를 토대로 하여 알 수 있는
사실을 하나 이상 제시하라

```python
from hossam import load_data
from matplotlib import pyplot as plt
from matplotlib import font_manager as fm
import seaborn as sb
import numpy as np
from pandas import DataFrame
```

```python
my_dpi = 200                                    # 이미지 선명도(100~300)
font_path = "./NotoSansKR-Regular.ttf"          # 한글을 지원하는 폰트 파일의 경로
fm.fontManager.addfont(font_path)               # 폰트의 글꼴을 시스템에 등록함
font_prop = fm.FontProperties(fname=font_path)  # 폰트의 속성을 읽어옴
font_name = font_prop.get_name()                # 읽어온 속성에서 폰트의 이름만 추출     
plt.rcParams['font.family'] = font_name         # 그래프에 한글 폰트 적용
plt.rcParams['font.size'] = 10                  # 기본 폰트크기
plt.rcParams['axes.unicode_minus'] = False      # 그래프에 마이너스 깨짐 방지(한글환경에서 필수)
```

```python
origin = load_data('kings_life')
origin
```

```python
origin.describe()
```

```python
width_px = 800
height_px = 350
rows = 1
cols = 1
figsize = (width_px/ my_dpi, height_px/ my_dpi)
fig, ax = plt.subplots(rows, cols, figsize=figsize, dpi= my_dpi)

sb.boxplot(data=origin, x='수명')



ax.grid(True)


plt.tight_layout()
plt.show()
plt.close()
```

평균과 중앙값이 비슷하며 중앙값이 40-50 사이에 있어 대체로 수명이 짧은 편이다.

외국기업과 내국기업별로 기술통계량을 상자그림을 그려서 비교/분석하라. 최종적으로 분석 결과를 토대
로 하여 알 수 있는 사실을 하나 이상 제시하라

```python
origin = load_data('stock')
origin
```

```python
width_px = 800
height_px = 500
rows = 1
cols = 1
figsize = (width_px/ my_dpi, height_px/ my_dpi)
fig, ax = plt.subplots(rows, cols, figsize=figsize, dpi= my_dpi)

sb.boxplot(data=origin,x='구분',y='주가', palette='pastel')

ax.grid(True)


plt.tight_layout()
plt.show()
plt.close()
```

외국기업의 중앙값과 하한값의 간격이 더 크고 150이 중앙값임을 알 수 있다.
국내 기업의 중앙값은 150보다 높으며 상한값도 외국 기업에 비해 더 높음.


비료 종류별로 기술통계량을 구하고 상자그림을 그려라. 최종적으로 분석 결과를 토대로 하여 알 수 있는
사실을 하나 이상 제시하라

```python
origin = load_data('grape')
origin
```

```python
df= DataFrame(origin.groupby('비료종류').describe().T)
df
```

```python
width_px = 1500
height_px = 700
rows = 1
cols = 1
figsize = (width_px/ my_dpi, height_px/ my_dpi)
fig, ax = plt.subplots(rows, cols, figsize=figsize, dpi= my_dpi)

sb.boxplot(data=df, palette='Set2')


ax.grid(True)

plt.tight_layout()
plt.show()
plt.close()
```

B는 중앙값이 20이하이고 중앙값으로부터 상한값과 하한값의 범위가 비슷하다.
A의 수확량의 상한값은 B보다 높으며 중앙값이 20이상이다.
A의 비료를 사용했을때 대체로 수확량이 더 많다는걸 알 수 있다.

```python
origin = load_data('analysis_grade')
origin
```

```python
df= DataFrame(origin.groupby('학과').describe().T)
df
```

```python
width_px = 1500
height_px = 700
rows = 1
cols = 1
figsize = (width_px/ my_dpi, height_px/ my_dpi)
fig, ax = plt.subplots(rows, cols, figsize=figsize, dpi= my_dpi)

sb.boxplot(data=df, palette='pastel')


ax.grid(True)

plt.tight_layout()
plt.show()
plt.close()
```

C가 중앙값과 평균이 A보다 더 높고 분산이 작다.
대체로 C학과가 더 점수가 높은걸 알 수 있다.

중간고사와 기말고사의 중앙값과 평균을 비교했을때  기말고사가 대체로 더 높음.

```python
width_px = 1280
height_px = 640

figsize = (width_px/ my_dpi, height_px/ my_dpi)
fig, ax = plt.subplots(1, 1, figsize=figsize, dpi= my_dpi)


sb.histplot(data=origin, x='중간고사',bins=[30,40,50,60,70,80,90,100],  color='skyblue')

# 기말고사 히스토그램
sb.histplot(data=origin, x='기말고사',bins=[30,40,50,60,70,80,90,100], color='salmon')

plt.tight_layout()
plt.show()
plt.close()
```

80점이상부터 중간고사보다 기말고사를 잘 본 친구들이 더 많다.
30-40점대는 중간고사가 기말고사보다 못 본 친구들이 더 많다.
