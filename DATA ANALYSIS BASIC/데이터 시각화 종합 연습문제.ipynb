문제 1.
수명에 대해 기술통계량을 구하고 상자그림을 그려라. 최종적으로 분석 결과를 토대로 하여 알 수 있는
사실을 하나 이상 제시하라

```python
from hossam import load_data
from matplotlib import pyplot as plt
from matplotlib import font_manager as fm
import seaborn as sb
import numpy as np
from pandas import DataFrame
```

```python
my_dpi = 200                                    # 이미지 선명도(100~300)
font_path = "./NotoSansKR-Regular.ttf"          # 한글을 지원하는 폰트 파일의 경로
fm.fontManager.addfont(font_path)               # 폰트의 글꼴을 시스템에 등록함
font_prop = fm.FontProperties(fname=font_path)  # 폰트의 속성을 읽어옴
font_name = font_prop.get_name()                # 읽어온 속성에서 폰트의 이름만 추출     
plt.rcParams['font.family'] = font_name         # 그래프에 한글 폰트 적용
plt.rcParams['font.size'] = 10                  # 기본 폰트크기
plt.rcParams['axes.unicode_minus'] = False      # 그래프에 마이너스 깨짐 방지(한글환경에서 필수)
```

```python
origin = load_data('kings_life')
origin
```

```python
origin.describe()
```

```python
width_px = 800
height_px = 350
rows = 1
cols = 1
figsize = (width_px/ my_dpi, height_px/ my_dpi)
fig, ax = plt.subplots(rows, cols, figsize=figsize, dpi= my_dpi)

sb.boxplot(data=origin, x='수명')



ax.grid(True)


plt.tight_layout()
plt.show()
plt.close()
```

평균과 중앙값의 차이가 크지 않고, 중앙값이 40~50세 구간에 위치해 전반적으로 수명이 짧은 편임을 알 수 있다.

2. 중간고사와 기말고사의 중앙값과 평균을 비교했을때  기말고사가 대체로 더 높음.

```python
width_px = 1280
height_px = 640

figsize = (width_px/ my_dpi, height_px/ my_dpi)
fig, ax = plt.subplots(1, 1, figsize=figsize, dpi= my_dpi)


sb.histplot(data=origin, x='중간고사',bins=[30,40,50,60,70,80,90,100],  color='skyblue')

# 기말고사 히스토그램
sb.histplot(data=origin, x='기말고사',bins=[30,40,50,60,70,80,90,100], color='salmon')

plt.tight_layout()
plt.show()
plt.close()
```

3.80점 이상 구간에서는 중간고사보다 기말고사를 더 잘 본 학생들이 더 많다.

반면 30~40점대에서는 기말고사보다 중간고사를 더 못 본 학생들이 더 많다.

```python
width_px = 1280
height_px = 640

figsize = (width_px/ my_dpi, height_px/ my_dpi)
fig, (ax1,ax2) = plt.subplots(1, 2, figsize=figsize, dpi= my_dpi)


sb.histplot(data=origin, x='중간고사',bins=[30,40,50,60,70,80,90,100],  color='skyblue',hue='전공',multiple='stack',ax=ax1)

sb.histplot(data=origin, x='기말고사',bins=[30,40,50,60,70,80,90,100], color='salmon',hue='전공',multiple='stack', ax=ax2)
plt.show()


plt.tight_layout()
plt.close()
```

4.80점 이상 구간에서는 중간고사보다 기말고사 점수가 더 높은 학생들이 많다.

또한 중간고사와 기말고사 모두에서 STAT 전공 학생들이 80~100점대를 더 많이 차지하고 있다.

문제3.
구동방식에 따른 자동차 수를 하나의 그래프로 시각화 하시오.

```python
width_px  = 1500
height_px = 640
figsize = (width_px / my_dpi, height_px / my_dpi)

fig, ax = plt.subplots(1, 1, figsize=figsize, dpi=my_dpi)

sb.countplot(data=origin,x='drv',edgecolor="#000000",linewidth=0.8,color='skyblue')

plt.xlabel("구동방식")
plt.ylabel("자동차 수")
plt.xticks()

plt.tight_layout()
plt.show()
plt.close()
```

4.자동차 구동 방식에 따른 도심 연비 평균을 하나의 그래프로 시각화 하시오.

```python
mean_cty = origin.groupby('drv')['cty'].mean().reset_index()

width_px  = 1500
height_px = 640
figsize = (width_px / my_dpi, height_px / my_dpi)

fig, ax = plt.subplots(1, 1, figsize=figsize, dpi=my_dpi)

sb.barplot(data=origin,x='drv',y='cty', edgecolor="#000000",linewidth=0.8,color='skyblue')

plt.xlabel("구동방식")
plt.ylabel("평균 도심 연비")
plt.xticks()

plt.tight_layout()
plt.show()
plt.close()
```
