```python
from pandas import DataFrame, read_csv, read_excel
```

```python
# 어느 학급의 성적표를 표현한 2차원 리스트
grade_data = [
                [1, "남자", 98, 77, 88, 64],
                [2, "여자", 88, 90, 62, 72],
                [1, "남자", 92, 70, 83, 79],
                [3, "여자", 63, 60, 31, 70],
                [4, "남자", 75, 50, 90, 88]
            ]
#2차원 리스트를 데이터 프레임으로 변환
# -> 학생별 점수 표현
df = DataFrame(grade_data)
df
```

```python
#인덱스 이름으로 사용할 리스트
i_names = ['철수', '영희', '민철', '수현', '호영']


# 칼럼 이름으로 사용할 리스트
c_names = ['학년', '성별', '국어', '영어', '수학', '과학']


#인덱스와 칼럼이름 지정하기
df = DataFrame(grade_data, index=i_names, columns=c_names)
df
```

```python
# 어느 학급의 성적표를 표현한 딕셔너리
grade_dic = {
    '학년': [1,2,1,3,4],
    '성별' : ['남자', '여자', '남자', '여자','남자'],
    '국어' : [98, 88, 92, 63, 75],
    '영어' : [77, 90, 70, 60, 50],
    '수학' : [88, 62, 83, 31, 90],
    '과학' : [64, 72, 79, 70, 88]
}

# 딕셔너리 key가 칼럼(열)이름이 된다.
df = DataFrame(grade_dic)
df
```

```python
#인덱스 지정하기
df = DataFrame(grade_dic,
               index=['철수', '영희', '민철', '수현', '호영'])

#인덱스 제목 지정하기
df.index.name = '이름'
df
```
