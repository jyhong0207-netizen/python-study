```python
from hossam import load_data
from pandas import CategoricalDtype
```

```python
origin = load_data("grade")
origin
```

```python
rows, cols = origin.shape
print("%d행, %d열의 %d차원 데이터" % (rows, cols, origin.ndim))
```

```python
origin.info()
```

```python
칼럼타입변경df = origin.copy()
칼럼타입변경df.info()
```

```python
칼럼타입변경df['국어'] = 칼럼타입변경df['국어'].astype('float')
칼럼타입변경df.info()
```

```python
타입변환df = origin.astype({'국어': 'float', '영어' : 'float',
                            '수학': 'str', '과학':'bool'})
타입변환df.info()
```

```python
origin.info()
```

```python
칼럼단위_명목df = origin.copy()
칼럼단위_명목df['학년'] = 칼럼단위_명목df['학년'].astype('category')
칼럼단위_명목df['성별'] = 칼럼단위_명목df['성별'].astype('category')
칼럼단위_명목df.info()
```

```python
범주형df =origin.astype({'학년': 'category', '성별': 'category'})
범주형df.info()
```

```python
컬럼단위_서열척도df = origin.copy()
컬럼단위_서열척도df['학년']= 컬럼단위_서열척도df['학년'].astype(
        CategoricalDtype(categories=[1,2,3,4], ordered=True))
컬럼단위_서열척도df.info()
```

```python
서열척도df= origin.astype({
    "학년":CategoricalDtype(categories=[1,2,3,4], ordered=True)
})
서열척도df.info()
```
