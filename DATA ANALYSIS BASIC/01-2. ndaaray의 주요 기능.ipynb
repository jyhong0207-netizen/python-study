```python
# 철수의 1학년~4학년까지의 평균 점수
grade = np.array([82,77,91,88])
grade
```

```python
# ->추출하고자 하는 원본과 같은 사이즈의 배열을 생성한다.
# -> 추출할 값은 True, 그렇지 않으면 False
bool_array = np.array([True, False, True, False])
bool_array
```

```python
# -> 조건에 맞는 항목만 1차 배열로 추출
result1 = grade[bool_array]
result1
```

```python
# 80점이상은 원소만 추출하기
# ->조건을 명시하기 위해 배열의 이름이 조건식에 명시된다.
mypoint = np.array([82,77,91,88])
ge80 = mypoint[mypoint>=80]
ge80
```

```python
ge80le90 = mypoint[np.logical_and(mypoint >=80, mypoint <=90)]
ge80le90
```

```python
gt_or_lt = mypoint[np.logical_or(mypoint<80, mypoint>90)]
gt_or_lt
```

```python
exam = np.array([60, 83, 75, 92, 86])

#첫 번째 파라미터로 주어진 조건이 참일 경우 '합격',
# 그렇지 않을 경우 '불합격'

result = np.where(exam>=80, '합격', '불합격')
result
```

```python
exam = np.array([60,85,51,79,94,88])

#학점을 부여하기 위한 점수의 구간을 설정하는 조건들을 리스트로 설정
conditions = [ (exam>=90), (exam >=80), (exam>70)]
                 #if         elif           elif

#조건에 따라 부여될 학점 
value = ['A','B', 'C']

#조건과 조건을 판단하기 위한 값을 지정함
# default는 조건에 맞는 값이 아닌 경우 사용됨
result = np.select(conditions,value,default='F')
result
```
