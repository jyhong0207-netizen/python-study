```python
from hossam import load_data
from pandas import concat
```

```python
강남점 = load_data('shop_강남')
강남점
```

```python
종로점 = load_data('shop_종로')
종로점
```

```python
강남점cp = 강남점.copy()
강남점cp['Price'] = 강남점cp['Price'].fillna(8000)
강남점cp
```

```python
종로점cp = 종로점.copy()
idx =  종로점cp[종로점cp['Product'] == 'Keyboard'].index[0]
종로점cp.loc[idx, 'Quantity']=9
종로점cp
```

```python
종로점cp = 종로점.copy()
종로점cp['Quantity'] = 종로점cp['Quantity'].replace(999, 9)
종로점cp
```

```python
df_total = concat([강남점cp, 종로점cp], ignore_index=True)
df_total
```
