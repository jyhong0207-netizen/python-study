```python
df = load_data("order_history")
df
```

```python
결측치처리 = df.copy()
mean_price = 결측치처리['price'].mean()
결측치처리['price'] = 결측치처리['price'].fillna(value=mean_price)
결측치처리
```

```python
이상치처리 = 결측치처리.copy()
median_price = 이상치처리['price'].median()
이상치처리.loc[이상치처리['price']> 1000, 'price']=median_price
이상치처리
```

```python
병합 = merge(이상치처리, df, on='customer_id')
병합
```

```python
타입변환 = 병합.astype({'customer_id': 'str'})
타입변환.describe()
```
