## #01. Series 자료형

### 라이브러리 참조


```python
import numpy as np
from pandas import Series
```

```python
#리스트를 통한 객체 생성
my_list = [10,30]
s1 = Series(my_list)
s1
```

```python
#ndarray를 통한 객체 생성
my_array = np.array([100,200])
s2=Series(my_array)
s2
```

```python
# 딕셔너리를 통한 객체 생성
my_dict = {"hello":123, "world": 456, "nice":789}
s3 = Series(my_dict)
s3
```

```python
my_series = Series({"hello":123, "world": 456,"nice" :789})
my_series.dtype
```

```python
my_series.info()
```

```python
my_series = Series([10,20,30,40,50])
print("==인덱싱==")
print(my_series[0])
print("==슬라이싱==")
print(my_series[1:3])
```

```python
my_series = Series({"a":100, "b":200, "c":300, "d":400})
print("==인덱싱==")
print(my_series["a"])
print("==슬라이싱==")
print(my_series["b":"d"])
```

```python
my_series = Series({"a":100, "b":200, "c":300, "d":400})

print(type(my_series.values))
my_series.values
```

```python
mylist = list(my_series.values)
mylist
```

```python
my_series = Series({"a":100,"b":200, "c":300, "d": 400})

print(type(my_series.index))
my_series.index
```

```python
mylist = list(my_series.index)
mylist
```
