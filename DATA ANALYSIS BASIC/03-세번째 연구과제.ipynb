```python
from hossam import load_data
from pandas import Series
```

```python
monsters = load_data("monsters")
monsters
```

```python
df1 = monsters.drop(columns=['last_seen'])
df1.head
```

```python
new_monster = Series({'monster_id': 'M021', 'height':2.2, 'height': 'Lake','species':'Kelpie',
                      'cautiion_level':'Mediup'})

df2 = df1.copy()
df2.loc[df2.index.max()+1]=new_monster
df2
```

```python
df3 =df2.copy()
m003_data = df3[df3['monster_id'] =='M003'].copy()
m003_data['monster_id']="M003_copy"

df3.iloc[df3[df3['monster_id']== 'M004'].index] = m003_data
```
