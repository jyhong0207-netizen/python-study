```python
from hossam import load_data
from pandas import DataFrame
```

```python
origin = load_data('vip_customer')
origin
```

```python
r2 = origin.query('City == "서울"')
r2
```

```python
r3 = origin.query('TotalSpending >100000')
r3
```

```python
r4 = origin.query('City == "서울"and TotalSpending >100000' ) #우수 고객은 총 구매액이 100,000원 이상인 고객, 활동적인 고객은 서울시에 거주하는 고객
r4
```

```python
r5 = r4.sort_values('TotalSpending', ascending=False)
r5
```

```python
fl = r5.filter(['Name', 'City','TotalSpending'])
fl
```
