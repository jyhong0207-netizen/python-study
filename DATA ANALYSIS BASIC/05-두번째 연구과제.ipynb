```python
from hossam import load_data
from pandas import DataFrame
```

```python
df = load_data("game_scores")
df
```

```python
df1 = df.dropna()
df2= df1.set_index('player')
df3 = DataFrame(df2.stack())
df3
```

```python
df4 = df3.reset_index()
df4.filter(['player', 0]).groupby('player').mean()
```
