```python
from hossam import load_data
from pandas import CategoricalDtype
```

```python
origin = load_data("fruit_shop")
origin
```

```python
rows, cols = origin.shape
ndim = origin.ndim
print("%d %d열의 %d차원 데이터" % (rows,cols,ndim))
```

```python
exam_df2 = origin.rename(columns={"Num_Sold": "판매량"})
exam_df2
```

```python
exam_df3 = exam_df2.astype({
    '과일' : 'category',
    '등급' : CategoricalDtype(categories=['A','B','C'], ordered=True)
})

exam_df3.info()
```
