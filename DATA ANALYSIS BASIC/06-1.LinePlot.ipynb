```python
from hossam import load_data
#글꼴을 시스템에 등록
from matplotlib import font_manager as fm 
#캔버스(figure)를 생성, 기본그래픽 함수 제공
from matplotlib import pyplot as plt
#고급그래픽 기능 제공
import seaborn as sb
```

```python
font_path = "./NotoSansKR-Regular.ttf" #한글을 지원하는 폰트 파일의 경로
fm.fontManager.addfont(font_path)      #폰트의 글꼴을 시스템에 등록함
font_prop = fm.FontProperties(fname=font_path) #폰트의 속성을 읽어옴
font_name = font_prop.get_name() #읽어온 속성에서 폰트의 이름만 추출
font_name #글꼴 이름 확인
```

```python
my_dpi = 120                                # 이미지 선명도를 결정하는 1인치당 점(픽셀의 수)
my_font_name= "Noto Sans KR"                # 시스템에 등록된 글꼴이름
my_theme ="dark"                            # 그림스타일 지정
                                    
sb.set_theme(style=my_theme)                #seaborn 스타일(화풍 설정하기)

plt.rcParams['font.family'] = my_font_name  #그래프에 한글폰트 적용
plt.rcParams['font.size']=16                #기본 폰트 크기
plt.rcParams['axes.unicode_minus']=False    #그래프에 마이너스 깨짐 방지(한글환경에서 필수)
```

```python
# 1)그래프 초기화(캔버스(fig)와 도화지(ax)준비하기)
width_px = 1700 
height_px =800
rows = 1
cols=1
figsize = (width_px /my_dpi, height_px/ my_dpi )
fig, ax = plt.subplots(rows, cols, figsize=figsize, dpi=my_dpi)
             
ax.set_title("제목", color ="#156FE6", fontsize=30,
             fontweight=700, pad=20)
ax.set_xlabel("X축", color="#78925F", fontsize=16, labelpad=10)
ax.set_ylabel("Y축", color="#EE8834", fontsize=16, labelpad=20)

plt.tight_layout()
#plt.savefig("myplot.png", dpi=my_dpi)
plt.show()
plt.close()
```
