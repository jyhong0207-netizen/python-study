```python
from hossam import load_data
from pandas import pivot_table,melt
```

```python
df=load_data("icecream_sales")
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df2=df.dropna(subset=['Quantity'])
df2
```

```python
df3 = pivot_table(df2, index = 'Date', columns='Flavor', values= 'Quantity')
df3
```

```python
df4 = df3.reset_index()
df4
```

```python
melt(df4, id_vars=['Date'], value_vars=['딸기','민트초코', '바닐라','초콜릿'],
     var_name = 'Flavor', value_name='Quantity')
```
