```python
from hossam import load_data
from sklearn.preprocessing import MinMaxScaler,MaxAbsScaler, StandardScaler, RobustScaler
from pandas import DataFrame
```

```python
origin= load_data('performance')
origin
```

```python
minmax_df = origin

x = minmax_df['MonthlySales']
xmin = x.min()
xmax = x.max()

minmax_df['MonthlySales'] = (x-xmin)/(xmax-xmin)
minmax_df.head()
```

```python
abs_df = origin
x =abs_df['CustomerSatisfaction'] = x/x.abs().max()
abs_df.head()
```

```python
std_df = origin
x = origin['ProjectHours'] 
std_df['ProjectHours'] = (x-x.mean())/x.std()
std_df.head()
```

```python
std_df = origin
x = origin['InnovationScore'] 
std_df['InnovationScore'] = (x-x.mean())/x.std()
std_df.head()
```
