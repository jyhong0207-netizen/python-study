```python
from hossam import load_data
```

```python
origin = load_data('group_list')
origin
```

```python
# 원본 데이터 프레임 복사
df = origin.copy()

df['이메일_소문자'] = df['이메일'].str.lower()
df['이메일_대문자'] = df['이메일'].str.upper()
df['단어_첫글자'] = df['아이디'].str.title()        #단어 첫 글자 대문자
df['첫글자'] = df['아이디'].str.capitalize()        #문장 첫 글자 대문자
df[['이메일', '이메일_소문자', '이메일_대문자', '단어_첫글자', '첫글자']].head()
```

```python
df['strip'] = df['메모'].str.strip()
df['lstrip']= df['메모'].str.lstrip()
df['rstrip'] = df['메모'].str.rstrip()
df['zfill']= df['아이디'].astype(str).str.zfill(12)
df['pad'] = df['아이디'].astype(str).str.pad(12, side='both',fillchar = '*' )
df[['strip', 'lstrip', 'rstrip', 'zfill', 'pad']].head()
```

```python
# 이메일이 'naver.com' 을 포함하는지 확인
df['네이버이메일'] = df['이메일'].str.contains('naver.com')
# 전화번호가 '010-'으로 시작하는지
df['휴대폰번호'] = df['전화번호'].str.startswith('010-')
# 주소가 '구'로 끝나는지
df['구단위주소'] = df['주소'].str.endswith('구')
#결과 확인
df[['이메일', '네이버이메일', '전화번호', '휴대폰번호', '주소', '구단위주소']].head()
```

```python
#메모에서 '중' 문자가 처음 나오는 위치(0부터 시작, 못찾으면 -1)

df['메모_중_위치'] = df['메모'].str.find('중')

# 메모에서 '중'이라는  글자의 개수
df['메모_중_개수'] = df['메모'].str.count('중')
# '메모 없음'이라는 문자열을 'empty'로 치환
df['메모없음'] = df['메모'].str.replace('메모없음', 'empty')
df[['메모', '메모_중_위치','메모_중_개수', '메모없음']].head()
```

```python
#'@'기준으로 분할한 후 0번째요소 반환
df['사용자명'] = df['이메일'].str.split('@').str.get(0)
#'@'기준으로 분할한 수 1번째 요소 반환
df['도메인']= df['이메일'].str.split('@').str.get(1)
#이름과 이메일 뒷부분 결합
df['이름_이메일'] = df.index.str.cat(df['도메인'], sep='@')
df[['사용자명', '도메인', '이름_이메일']].head()
```

```python
df['아이디_숫자여부(1)'] = df['아이디'].str.isdigit()
df['아이디_숫자여부(2)'] = df['아이디'].str.isnumeric()
df['등급_문자여부'] = df['등급'].str.isalpha()
df['등급_문자+숫자여부'] = df['등급'].str.isalnum()

df[['아이디', '아이디_숫자여부(1)', '아이디_숫자여부(2)', '등급', '등급_문자여부', '등급_문자+숫자여부']].head()
```

```python
df['이름_길이'] = df.index.str.len()
df['이름_성'] = df.index.str.slice(0,1)
df['전화번호_뒷4자리'] = df['전화번호'].str.slice(-4)
df[['이름_길이', '이름_성', '전화번호', '전화번호_뒷4자리']].head()
```
